```python
import pandas as pd
%matplotlib inline
import seaborn as sns
import numpy as np
import matplotlib.pyplot as plt

```

```python
data = pd.read_csv("data/data_numerical.csv")
data = data.replace("?", np.nan)
```

```python
cols_with_missing = data.columns[data.isnull().any()]

# Print columns with missing values
print("Columns with missing values:")
print(cols_with_missing)

for col in cols_with_missing:
    plt.figure(figsize=(8, 6))
    sns.histplot(data[col].dropna(), kde=True, bins=30)
    plt.title(f"Distribution of {col}")
    plt.xlabel(col)
    plt.ylabel("Density")
    
    # Customize x-axis ticks for better visualization
    plt.xticks(rotation=45, ha='right')  # Rotate labels for better readability
    plt.tight_layout()  # Adjust layout to prevent overlapping
    
    plt.show()

```

```python
# Calculate skewness for each numeric column
skewness = data.select_dtypes(include='number').apply(lambda x: x.skew())

# Fill missing values in numeric columns based on skewness
for col in skewness.index:
    if skewness[col] > 1 or skewness[col] < -1:
        # Right-skewed or left-skewed, fill with median
        data[col].fillna(data[col].median(), inplace=True)
    else:
        # Normally distributed, fill with mean
        data[col].fillna(data[col].mean(), inplace=True)

# Fill missing values in categorical columns with mode
categorical_cols = data.select_dtypes(include='object').columns
data[categorical_cols] = data[categorical_cols].fillna(data[categorical_cols].mode().iloc[0])
```

```python
import plotly.express as px

# Assuming 'df' is your DataFrame containing the 'Gender' column
labels = data['sex'].value_counts().index
values = data['sex'].value_counts().values

# Define colors for the pie chart
colors = ['#9cc359', '#e96b5c']

# Create the pie chart figure using Plotly Express
fig = px.pie(data, values=values, names=labels, title='Patients by Gender', color_discrete_sequence=colors)

# Show the plot
fig.show()
```

```python
import plotly.express as px

# Age distribution bar plot
age_counts = data['age'].value_counts().reset_index().rename(columns={'index': 'Age', 'age': 'Count'})
fig_age = px.bar(age_counts, x='Age', y='Count', title='Age Distribution of Patients')

# Duration of Parkinson's disease (durat_pd) bar plot
durat_pd_counts = data['durat_pd'].value_counts().reset_index().rename(columns={'index': 'Duration (years)', 'durat_pd': 'Count'})
fig_durat_pd = px.bar(durat_pd_counts, x='Duration (years)', y='Count', title='Duration of Parkinson\'s Disease')

# Show the plots
fig_age.show()
fig_durat_pd.show()
```

```python
from sklearn.ensemble import RandomForestClassifier
import matplotlib.pyplot as plt
from sklearn.ensemble import ExtraTreesClassifier

# Assuming 'data' is your DataFrame containing the features and target variable
# Assuming 'target_column' is the name of your target column
# Replace 'target_column' with your actual target column name

# Separate features (X) and target variable (y)
X = data.drop(columns=['hy','patnum'])  # Independent columns
X = pd.get_dummies(X)
# Specify the target column (y)
y = data['hy'] 

# Train the ExtraTreesClassifier model
model = ExtraTreesClassifier()
model.fit(X, y)

# Print feature importances
print(model.feature_importances_)

# Plot graph of feature importances for better visualization
feat_importances = pd.Series(model.feature_importances_, index=X.columns)
feat_importances.nlargest(20).plot(kind='barh')  # Plot top 10 most important features
plt.xlabel('Feature Importance Score')
plt.ylabel('Features')
plt.title('Top 10 Most Important Features')
plt.show()
```

```python
# Iterate through each column
for column in data.columns:
    print(f"Attribute: {column}") 
    print("------------")
    
    # Check data type
    print("Data Type:", data[column].dtype)
    
    # Check for missing values
    print("Missing Values:", data[column].isnull().sum())
    
    # Display summary statistics
    print("Summary Statistics:")
    print(data[column].describe())
    
    # Visualize the distribution (example: histogram for numerical data)
    if data[column].dtype != 'object':
        data[column].hist()
        plt.xlabel(column)
        plt.ylabel('Frequency')
        plt.title(f'Distribution of {column}')
        plt.show()
    
    # Visualize categorical data (example: bar plot for categorical data)
    if data[column].dtype == 'object':
        data[column].value_counts().plot(kind='bar')
        plt.xlabel(column)
        plt.ylabel('Count')
        plt.title(f'Count of {column}')
        plt.show()
    
    print("\n")
```
